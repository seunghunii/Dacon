```python
breakfast = df['조식메뉴'].str.replace('\(.*\)','',regex=True).str.strip()
breakfast[0:5]
```

```python
regex_to_split = re.compile(' {1,}|/')
breakfast_split = [regex_to_split.split(x) for x in breakfast]
```

```python
breakfast_tmp = []
for k in range(len(breakfast_split)):
    breakfast_tmp += breakfast_split[k]
```
