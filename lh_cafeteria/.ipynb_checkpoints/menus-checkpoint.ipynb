```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data_dir = 'C:/Users/rsh15/Desktop/seunghuni/dataset/lh_cafeteria_dataset'
```

```python
df = pd.read_csv(data_dir+'/train.csv')
df.head()
```

```python
def tokenize(text):
    return text.split(',')

def menu_tokenizer(df,colname):
    menu_strs = df[colname].str.replace('\(.*\)','',regex=True)
    menu_strs = menu_strs.str.replace('<','').str.replace('>','')
    menu_strs = menu_strs.str.replace('/ ','/').str.replace(' /','/').str.strip()
    
    regex_to_split = re.compile(' {1,}|/')
    menu_to_split = [regex_to_split.split(x) for x in menu_strs]
    
    tmpp = [','.join(x) for x in menu_to_split]
    
    vectorizer = CountVectorizer(tokenizer=tokenize)
    a = vectorizer.fit_transform(tmpp)
    menus = vectorizer.get_feature_names()
    menu_df = pd.DataFrame(a.toarray(),columns = menus)
    menu_df.columns = [colname + '_' + x for x in menu_df.columns.values]
    return menu_df
    
```

```python
breakfast = menu_tokenizer(df,'조식메뉴')
breakfast.head()
```
