## hak(school, len)

```python
school_list = []
for k in hakk:
    k_list = [j.strip() for j in k.splitlines() if not k == '']
    
```

```python
def assembly_hak_list(hak_str):
    reg_school = re.compile('대학|대학교|(.*?)대 (.*?)과|졸업$|!명예')
    stage_a = [k.strip() for k in hak_str.splitlines() if not k == '']
    hak_len.append(len(stage_a))
    
    stage_c = [i for i in stage_a if reg_school.search(i)]
    hak_sch.append(stage_c)
    hak_sch_len.append(len(stage_c))
```

```python
hak_len = []
hak_sch = []
hak_sch_len = []
for d in hakk:
    assembly_hak_list(d)
```

```python
fig,ax = plt.subplots(2,1,figsize=(20,8))
sns.countplot(hak_sch_len,ax=ax[0])
sns.countplot(hak_len,ax=ax[1])
```

```python
fig,ax = plt.subplots(2,1,figsize=(20,8))
sns.barplot(people['DAE_NM'],hak_sch_len,estimator=np.mean,ci=False,ax=ax[0])
sns.barplot(people['DAE_NM'],hak_len,estimator=np.mean,ci=False,ax=ax[1])
```

```python
plt.figure(figsize=(15,8))
sns.barplot(people['DAE_NM'],hak_len,estimator=np.mean,ci=False,color='blue')
sns.barplot(people['DAE_NM'],hak_sch_len,estimator=np.mean,ci=False,color='black')
```
