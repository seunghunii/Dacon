## data load

```python
cus_info = pd.read_csv('C:/Users/rsh15/Desktop/seunghuni/dataset/리그2/2_cus_info.csv')
act_info = pd.read_csv('C:/Users/rsh15/Desktop/seunghuni/dataset/리그2/2_act_info.csv')
trd_kr   = pd.read_csv('C:/Users/rsh15/Desktop/seunghuni/dataset/리그2/2_trd_kr.csv')
trd_os   = pd.read_csv('C:/Users/rsh15/Desktop/seunghuni/dataset/리그2/2_trd_oss.csv')
iem_info = pd.read_csv('C:/Users/rsh15/Desktop/seunghuni/dataset/리그2/2_iem_info.csv')
```

```python
trd_kr['iem_cd'] = [x.strip() for x in trd_kr['iem_cd']]
iem_info['iem_cd'] = [x.strip() for x in iem_info['iem_cd']]
iem_info['iem_krl_nm'] = [x.strip() for x in iem_info['iem_krl_nm']]
```

```python
trd_kr['orr_dt'] = pd.to_datetime(trd_kr['orr_dt'],format='%Y%m%d')
trd_os['orr_dt'] = pd.to_datetime(trd_os['orr_dt'],format='%Y%m%d')
```

```python
# 날짜가 0인 데이터가 있음. 2개.
# 이 두 계좌는 국내 해외 둘 다 주식거래가 없으므로 drop 한다.
# drop 후 datetime 날짜 변환
act_empty = act_info[act_info['act_opn_ym'] == 0]['act_id']
trd_kr['act_id'].isin(act_empty).sum()
trd_os['act_id'].isin(act_empty).sum()

act_info.drop(act_info[act_info['act_opn_ym'] == 0].index,inplace=True)
act_info['act_opn_ym'] = pd.to_datetime(act_info['act_opn_ym'],format='%Y%m')
```

```python
act_cus = pd.merge(act_info,cus_info,how='left',
                   left_on='cus_id',right_on='cus_id')
```

# join korean data

```python
korea_df = pd.merge(trd_kr,act_cus,left_on = 'act_id',right_on = 'act_id',how='left')
```

```python
korea_df['act_id'] = [x[0:10] for x in korea_df['act_id']]
korea_df['cus_id'] = [x[0:10] for x in korea_df['cus_id']]
```
