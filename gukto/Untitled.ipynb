```python
# actual spent without dispent
df['actual_spent'] = df['totalspent'] - df['disspent']
df['numofactspent'] = df['numofspent'] - df['numofdisspent']
```

### point(x,y)

```python
proj_UTMK = Proj(init='epsg:5178')
proj_WGS84 = Proj(init='epsg:4326')
a,b = transform(proj_UTMK, proj_WGS84,df['point_x'].values,df['point_y'].values)
```

* 세부적으로 나누어보면, 월별 Actual Spent는 증가한 반면, Disspent는 감소했다. 재난지원금을 수령한 뒤 계속해서 소진하는 양상이다.

```python
fig,ax = plt.subplots(2,1,figsize=(18,9));
tmp_df = df.groupby('ym',as_index=False)['actual_spent','disspent'].sum()
sns.lineplot(data=tmp_df,x='ym',y='actual_spent',ax=ax[0]);
ax[0].set_title('Monthly Actual Spent');
ax[0].set_xticks([5,6,7,8]);

sns.lineplot(data=tmp_df,x='ym',y='disspent',ax=ax[1]);
ax[1].set_title('Monthly Dis Spent');
ax[1].set_xticks([5,6,7,8]);
```

* 월별 전체 소비 비율차이를 보면, 숙박/교통의 소비가 계속해서 증가했다.

```python
# 비율차이. 대부분 감소하거나 살짝 증가한 반면, 숙박/교통의 증가 비율은 좀 크다.
ex_df = df.groupby(['ym','tpbuz_cd1'],as_index=False)['totalspent'].sum()
ex_df = pd.pivot(ex_df,index='ym',columns='tpbuz_cd1').reset_index()
ex_df.columns = ['ym','가정','교육,문구','상품','서비스','숙박/교통','음식','의료/약품','차량','취미']
ex_df.drop('ym',axis=1).pct_change().drop(0,axis=0).plot.bar(figsize=(18,8));
plt.title('월별 전체 소비 비율차이',fontsize=20);
plt.xticks([0,1,2],['5-6','6-7','7-8'],fontsize=13,rotation=10);
```
