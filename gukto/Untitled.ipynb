```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
mpl.rcParams['font.family'] = 'JEJUGOTHIC'
```

```python
df_names = glob.glob('C:/Users/rsh15/Desktop/seunghuni/dataset/gukto/*.txt')
```

```python
df_list = []
for k in df_names:
    df_list.append(pd.read_csv(k))
```

```python
df_list[2].drop(['X','Y'],axis=1,inplace=True)
```

```python
for a in df_list:
    print(a.shape)
```

```python
df = pd.concat(df_list,axis=0)
df.columns = df.columns.str.lower()
df.columns
```

```python
df['ym'] = pd.to_datetime(df['ym'],format='%Y%m')
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
plt.figure(figsize=(18,8))
sns.lineplot(data=df,x='ym',y='totalspent',hue='franclass',ci=False)
plt.xticks(['2020-0'+str(x)+'-01' for x in range(5,9)],
           ['2020-0'+str(x) for x in range(5,9)])
```

```python
plt.figure(figsize=(18,8))
sns.relplot(data=df,x='ym',y='totalspent',hue='type',row='franclass',legend=False,kind='line')
```
