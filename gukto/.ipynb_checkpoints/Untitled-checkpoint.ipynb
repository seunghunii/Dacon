## basic preprocessing

### ym

```python
df['ym'] = df['ym'].replace({202005:5,202006:6,202007:7,202008:8})
```

### spent data

```python
# actual spent without dispent
df['actual_spent'] = df['totalspent'] - df['disspent']
df['numofactspent'] = df['numofspent'] - df['numofdisspent']
```

### time

```python
df.drop(df[df['time'] == 'x시'].index,inplace=True)
df['time'] = df['time'].apply(lambda x:x.replace('시','')).astype(int)
```

### point(x,y)

```python
proj_UTMK = Proj(init='epsg:5178')
proj_WGS84 = Proj(init='epsg:4326')
a,b = transform(proj_UTMK, proj_WGS84,df['point_x'].values,df['point_y'].values)
```

```python
df['point_x'] = a
df['point_y'] = b
```

##### 매출규모 등급에 따른 총 소비금액
#####  - 매출액 규모는 일반 - 영세 - 중소2 - 중소1 - 중소 순서.
#####  - 중간 등급인 매장들의 소비가 적다.

```python
plt.figure(figsize=(15,7))
df.groupby(['franclass'],as_index=True)['totalspent'].sum().plot.pie(y='totalspent',autopct='%.1f%%')
plt.ylabel('')
plt.title('규모별 Total Spent')
```

```python
plt.figure(figsize=(20,8))
tmp_df = df.groupby(['franclass','tpbuz_cd1'],as_index=False)['totalspent'].sum()
fig,ax = plt.subplots(2,3,figsize=(18,10))
pos = [(0,0),(0,1),(0,2),(1,0),(1,1),(1,2)]

for k in range(0,5):
    idxx = tmp_df['franclass'].unique()[k]
    ax[pos[k]].pie(tmp_df.loc[tmp_df['franclass']==idxx,'totalspent'],
                labels=tmp_df.loc[tmp_df['franclass']==idxx,'tpbuz_cd1'])
    ax[pos[k]].set_title(idxx)
fig.delaxes(ax[pos[5]])
```

##### 규모별 Plot을 봤을 때, 음식과 상품이 가장 많은 것을 알 수 있다.
##### 음식, 상품의 중분류별 총 소비를 살펴보면,
##### 음식은 일반음식, 서양음식, 중국음식 등이 있는 카테고리가,
##### 상품은 편의점, 인터넷몰, 대형할인점, 슈퍼마켓 등이 있는 카테고리가 대부분의 비율을 차지하고 있는 것을 볼 수 있다.

```python
fig,ax = plt.subplots(1,2,figsize=(15,7))
df[df['tpbuz_cd1']=='음식'].groupby(['tpbuz_cd2'],as_index=True)['totalspent'].sum().plot.pie(
    y='totalspent',autopct='%.1f%%',ax=ax[0])
ax[0].set_ylabel('')
ax[0].set_title('음식 중분류')

df[df['tpbuz_cd1']=='상품'].groupby(['tpbuz_cd2'],as_index=True)['totalspent'].sum().plot.pie(
    y='totalspent',autopct='%.1f%%',ax=ax[1])
ax[1].set_ylabel('')
ax[1].set_title('상품 중분류')
```
